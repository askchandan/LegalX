```python
import os
# Set your OpenAI API key as an environment variable before running this notebook
# You can set it in your terminal: set OPENAI_API_KEY=your_actual_key_here
# Or create a .env file and load it with python-dotenv

from langchain_community.document_loaders.csv_loader import CSVLoader

file_path = "ipc_sections.csv"

loader = CSVLoader(file_path=file_path)
data = loader.load()

for record in data[:2]:
    print(record)
```

```python
# %pip install -qU langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(data)
```

```python
print(splits[0])
```

```python
print(len(splits))
```

```python
from langchain_community.embeddings.huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

# You can specify a model name, e.g., "all-MiniLM-L6-v2"
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)
```

```python
print(vectorstore._collection.count())
```

```python
print(vectorstore._collection.get())
```

```python
print("\n collection 1 - ", vectorstore._collection.get(ids=['b59f4d94-fa1e-4279-a093-6c3c96c89079'], include=["documents", "embeddings"]))
```

```python
retriever = vectorstore.as_retriever()
```

```python
from langchain import hub
prompt = hub.pull("rlm/rag-prompt")
```

```python
from langchain_community.chat_models import ChatOllama
llm = ChatOllama(model='llama3.2')
```

```python
# whenever we need to pass input as it is, we use runnable pass-through
from langchain_core.runnables import RunnablePassthrough
# when we take the llm output , there is too much info., so we use output parser to take only the content
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(docs):
    return "\n".join([doc.page_content for doc in docs])
```

```python
rag_chain = ({"context": retriever | format_docs, "question": RunnablePassthrough()}
             |prompt
             |llm
             |StrOutputParser())
```

```python
rag_chain.invoke("muder punishmemnt in jail along with section?")
```
